```python
import matplotlib.pyplot as plt

Modelos = [
    "LSTM",
    "LSTM pre-trained embeddings",
    "GPT-2 instruct tuned",
    "Llama 3.2 3B",
    "Gemma 3 4B fine-tuned",
    "Llama 3.2 3B fine-tuning",
    "Gemma 3 4B"
]

Parametros = [
    70e6,  # 70M
    70e6,  # 70M
    124e6, # 1774M
    3e9,   # 3B
    4e9,   # 4B
    3e9,   # 3B
    4e9    # 4B
]

GLEU = [
    0.0549123762,
    0.08061482,
    0.1994957,
    0.5533238,
    0.61655266,
    0.60596251,
    0.5811729628
]

# Convertir el nmero de parmetros a billones
Parametros_billones = [p / 1e9 for p in Parametros]

# Asignar colores a cada tipo de modelo
colores = {
    "LSTM": "red",
    "GPT-2": "green",
    "Llama": "blue",
    "Gemma": "purple"
}

# Asignar un color a cada modelo
color_puntos = [
    colores["LSTM"],
    colores["LSTM"],
    colores["GPT-2"],
    colores["Llama"],
    colores["Gemma"],
    colores["Llama"],
    colores["Gemma"]
]

# Crear el scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(Parametros_billones, GLEU, color=color_puntos)

# Aadir etiquetas a los puntos, ajustando la posicin de las etiquetas para LSTM y otros modelos
for i, modelo in enumerate(Modelos):
    if "LSTM" in modelo:
        x_offset = 0.05
        ha = 'left'
    elif "GPT-2" in modelo:
        x_offset = 0.05
        ha = 'left'
    else:
        x_offset = -0.05
        ha = 'right'

    plt.text(Parametros_billones[i] + x_offset, GLEU[i], modelo, fontsize=8, ha=ha)

# Etiquetas de los ejes
plt.xlabel('Number of parameters (B)')
plt.ylabel('GLEU')

# Mostrar el grfico
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos organizados
Modelos = ["Gemma 3 4B fine-tuned", "Llama 3.2 3B fine-tuned"]
Pronombres = [0.7816516404, 0.6827980438]
GeneroYNumero = [0.7645415091, 0.6806682164]
Verbo = [0.736919, 0.7022648667]

# Configurar la posicin de las barras
x = np.arange(len(Modelos))  # Posiciones para los labels de los modelos
width = 0.25  # Ancho de cada barra

# Obtener los colores de la paleta tab10
colors = plt.get_cmap('tab10').colors

# Crear el grfico de barras usando la paleta tab10
fig, ax = plt.subplots(figsize=(12, 8))
rects1 = ax.bar(x - width, Pronombres, width, label='Pronouns and articles', color=colors[0])
rects2 = ax.bar(x, GeneroYNumero, width, label='Gender and number', color=colors[1])
rects3 = ax.bar(x + width, Verbo, width, label='Verbs', color=colors[2])

# Aadir ttulos y etiquetas
ax.set_xlabel('Model')
ax.set_ylabel('GLEU')

ax.set_xticks(x)
ax.set_xticklabels(Modelos)
ax.legend()

# Aadir anotaciones a las barras
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

# Ajustar diseo
fig.tight_layout()

# Mostrar el grfico
plt.show()
```
